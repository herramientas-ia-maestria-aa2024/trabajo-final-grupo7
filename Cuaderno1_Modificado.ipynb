```python
import pandas as pd
import pymysql
import sqlalchemy as db
```

```python
from faker import Faker
fake = Faker()
```

```python
lista = []
```

```python
#Se crea una lista con Faker de 1000 registros y 5 atributos, a manera de ejemplo
for x in range(0,1000):
    lista.append({'Nombre': fake.first_name(),'Apellido':fake.last_name(),'Dirección':fake.address(),'Correo electrónico':fake.email(),'Fecha de nacimiento':fake.date_of_birth()})
```

```python
lista
```

```python
#Se crea un dataframe de la lista anterior
data = pd.DataFrame(lista)
```

```python
#Se cambia el índice de 0 a 100 por 1 a 1000
rango=range(1,len(data)+1)
data.index=rango
```

```python
#Se imprime el dataframe data
data
```

```python
#Se crea una funcion para generar numeros de cedula (a manera de ejemplo)
cedula=1102792826
def generar_cedula(d):
    global cedula
    cedula+=1
    return cedula
```

```python
#Se añade la columna Cedula al dataframe data
data['Cédula']=data['Nombre'].apply(generar_cedula)
```

```python
#Se mueve la columna Cedula al inicio del dataframe data
data = data[['Cédula'] + [col for col in data.columns if col != 'Cédula']]
```

```python
#Se imprime el dataframe data
data
```

```python
#Se transforma en archivo csv el dataframe data
data.to_csv("data.csv", index=False)
```

```python
#Se crean dos dataframes a partir del dataframe data, mediante la eliminación de columnas
data_parte1=data.drop(columns=["Correo electrónico","Fecha de nacimiento"],inplace=False)
data_parte2=data.drop(columns=["Nombre","Apellido","Dirección"],inplace=False)
```

```python
#Se transforman en archivo csv a los dataframes data_parte1 y data_parte2
data_parte1.to_csv("data_parte1.csv", index=False)
data_parte2.to_csv("data_parte2.csv", index=False)
```
