```python
import pandas as pd
import mysql.connector
```

```python
#Lectura del primer dataframe
#Se lee el archivo csv data_parte1 y se le asigna a la variable data_parte1
data_parte1 = pd.read_csv("data_parte1.csv")
rango=range(1,len(data_parte1)+1)
#Se modifica el indice de 1 a 1000
data_parte1.index=rango
#Se imprime el dataframe data_parte1
data_parte1
```

```python
#Lectura del segundo dataframe
#Se crea la conexion con la base de datos creada en MySQL.La base de datos se llama 'datasets'
conexion=mysql.connector.connect(user='root', password='26061998santiago', host='localhost', database='datasets', port='3306')
#Se imprime la conecion para verificar que efectivamente se estableció
print(conexion)
```

```python
#Se crea un cursor para poder acceder a las tablas de la base de datos
cursor=conexion.cursor()
```

```python
#Se accede a la informacion de la tabla 'data_parte2_sql' creada en la base de datos 'datasets' en MySQL
sql="""SELECT * FROM data_parte2_sql"""
cursor.execute(sql)
```

```python
#Se obtiene el nombre de las columnas de la tabla 'data_parte2_sql'
columnas=cursor.column_names
#Se imprimen los nombres de las columnas
columnas
```

```python
#Se obtienen todos los registros de la tabla 'data_parte2_sql'
conjunto_datos=cursor.fetchall()
#Se imprimen los registros
conjunto_datos
```

```python
#Se transforma la lista de tuplas en lista de listas
lista_conjunto_datos=[]
for fila in conjunto_datos:
    lista_conjunto_datos.append(list(fila))
#Se imprime la lista de listas
lista_conjunto_datos                 
```

```python
#Se transforma en dataframe la lista de listas: 'lista_conjunto_datos'
data_parte2_ind=pd.DataFrame(lista_conjunto_datos)
data_parte2_ind.columns=columnas   #Se asigna el nombre de las columnas
data_parte2_ind.index=rango        #Se modifica el indice de 1 a 1000
data_parte2_ind                    #Se imprime el dataframe
```

```python
#Se elimina la columna 'index'
data_parte2=data_parte2_ind.drop(columns=["index"],inplace=False)
data_parte2   #Se imprime el dataframe
```

```python
#Se procede a unir los dataframes mediante el metodo merge, utilizando como columna en común a 'Cédula'
dataframe_unido=pd.merge(data_parte1,data_parte2, on='Cédula')
dataframe_unido.index=rango     #Se modifica el indice de 1 a 1000
dataframe_unido                 #Se imprime el dataframe unido
```

```python
#Se finaliza la conexion con la base de datos en MySQL
conexion.close()
```
